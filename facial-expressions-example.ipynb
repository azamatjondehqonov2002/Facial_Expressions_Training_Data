## Imports

```python
# data management
import pandas as pd

# read and show images
from cv2 import cv2
import matplotlib.pyplot as plt  # plot

# file management
import os
from os.path import join
```

## Viewing the Data

```python
path = ('/kaggle/input/affectnet-training-data/')
file = (path + 'labels.csv')
df = pd.read_csv(file)
df.head()
```

```python
fig, axs = plt.subplots(2, 2, sharey=True, constrained_layout=True, num=None, figsize=(5, 5), dpi=80, facecolor='gray', edgecolor='k')
fig.suptitle("Sample Faces and Labels")
axs = axs.flatten()

for i in range(4):
  img = (cv2.imread(path + df['pth'][i+2000]))  # read image
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # convert to BGR to RGB
  axs[i].imshow(img)
  axs[i].set_title(df['label'][i])
```
